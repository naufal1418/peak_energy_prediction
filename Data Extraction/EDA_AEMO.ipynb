```python
import os
import glob
import pandas as pd
os.chdir("../Data/raw/AEMO/Operational-demand")
```

# Merged all operational data on one csv

```python
extension = 'csv'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
```

```python
#combine all files in the list
combined_csv = pd.concat([pd.read_csv(f) for f in all_filenames ])
combined_csv = combined_csv.drop(['Trading Date', 'Interval Number', 'Extracted At'], axis = 1)
#export to csv
combined_csv.to_csv( "combined_csv-operational-demand.csv", index=False, encoding='utf-8-sig')
```

```python
df = pd.read_csv("combined_csv-operational-demand.csv", index_col='Trading Interval', parse_dates=['Trading Interval'])
df.head()
```
